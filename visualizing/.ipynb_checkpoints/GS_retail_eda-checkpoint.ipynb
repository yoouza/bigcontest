```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

path = 'C:/bg/bibibigcon/04_retail/'
```

```python
dong = pd.read_excel(path + '04_혁신아이디어분야_유통데이터(GS리테일)_데이터정의서 및 문제 데이터.xlsx',
                     header = 3,
                     sheet_name = 3)
category = pd.read_excel(path + '04_혁신아이디어분야_유통데이터(GS리테일)_데이터정의서 및 문제 데이터.xlsx',
                         sheet_name = 4)
```

```python
dong = dong.groupby(['ADMD']).sum()
date = category.columns[5:]
category = category.groupby(['ADMD','ANTC_ITEM_LCLS_NM'])[date].sum()
```

```python
dong
```

```python
category
```

```python
# 동 별 dong 데이터를 8번씩 반복한 numpy array 생성

amt_dong = np.zeros(category.shape)
for i, v in enumerate(np.array(dong)):
    amt_dong[8*i:8*(i+1)] = [v]*8
```

```python
# 곱셈 가능한 형태 확인
np.array(category).shape, amt_dong.shape
```

```python
amt = np.multiply(np.array(category), amt_dong)
df = pd.DataFrame(amt, index = category.index, columns = date)
df
```

```python
df_category = df.reset_index().groupby('ANTC_ITEM_LCLS_NM').sum().reset_index()
df_category
```

```python
df.isna().sum().sum()
```

```python
date
```

```python
date_2019 = date[:len(date)//2]
date_2020 = date[len(date)//2:]
```

# 시각화

```python
plt.figure(figsize=(20,10))
for i in range(len(df_category)):
    plt.plot(date_2019,
             df_category.drop('ANTC_ITEM_LCLS_NM', axis=1).iloc[:,:len(date_2019)].loc[i],
             label = df_category.ANTC_ITEM_LCLS_NM[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('2019 편의점 매출: 전체', size = 30)
plt.show()
```

```python
plt.figure(figsize=(20,10))
for i in range(len(df_category)):
    plt.plot(date_2020,
             df_category.drop('ANTC_ITEM_LCLS_NM', axis=1).iloc[:,len(date_2019):].loc[i],
             label = df_category.ANTC_ITEM_LCLS_NM[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('2020 편의점 매출: 전체', size = 30)
plt.show()
```

### 매출 스케일에 따른 클러스터링
- 큼(Cluster1): 간식, 마실거리, 식사, 헬스&뷰티
- 적음(Cluster2): 홈&리빙, 사회활동, 취미&여가활동, 임신/육아

```python
c1 = ['간식', '마실거리', '식사', '헬스&뷰티']
c2 = ['홈&리빙', '사회활동', '취미&여가활동', '임신/육아']
```

```python
df_c1 = df_category.groupby('ANTC_ITEM_LCLS_NM').sum().loc[c1]
df_c2 = df_category.groupby('ANTC_ITEM_LCLS_NM').sum().loc[c2]
```

```python
plt.figure(figsize=(20,10))
for i in range(len(c1)):
    plt.plot(date_2019,
             df_c1.iloc[:,:len(date_2019)].loc[c1[i]],
             label = c1[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('2019 편의점 매출: Cluster1', size = 30)
plt.show()
```

```python
plt.figure(figsize=(20,10))
for i in range(len(c2)):
    plt.plot(date_2019,
             df_c2.iloc[:,:len(date_2019)].loc[c2[i]],
             label = c2[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('2019 편의점 매출: Cluster2', size = 30)
plt.show()
```

```python
plt.figure(figsize=(20,10))
for i in range(len(c1)):
    plt.plot(date_2020,
             df_c1.iloc[:,len(date_2019):].loc[c1[i]],
             label = c1[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('2020 편의점 매출: Cluster1', size = 30)
plt.show()
```

```python
plt.figure(figsize=(20,10))
for i in range(len(c2)):
    plt.plot(date_2020,
             df_c2.iloc[:,:len(date_2020)].loc[c2[i]],
             label = c2[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('2020 편의점 매출: Cluster2', size = 30)
plt.show()
```

# 2019년과 2020년 비교
## 1. 요일에 맞춰서
- 처음에 버리는 데이터: 2019[0] (2019.02.01->버림)
- 2019년 [i + 1] == 2020년 [i]
- 마지막에 버리는 데이터: 2020[-2:] (2020.05.30/31 -> 버림)

```python
date_2019_diff = date_2019[1:]
date_2020_diff = date_2020[:-2]
```

```python
len(date_2019_diff), len(date_2020_diff)
```

```python
df_category.loc[:, date_2019_diff]
```

```python
diff = np.array(df_category.loc[:, date_2020_diff]) - np.array(df_category.loc[:, date_2019_diff])
diff = pd.DataFrame(diff, columns = date_2020_diff)
diff
```

```python
plt.figure(figsize=(20,10))
for i in range(len(diff)):
    plt.plot(date_2020_diff,
             diff.loc[i],
             label = df_category.ANTC_ITEM_LCLS_NM[i])
    plt.legend(fontsize = 'large')
    plt.xticks(rotation = 75, fontsize = 10)
    plt.title('작년 대비 2020년 매출 변동\n (기준: 요일)', size = 30)
plt.show()
```
