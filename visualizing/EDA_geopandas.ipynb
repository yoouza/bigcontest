```python
import geopandas as gpd
import pandas as pd
```

```python
dbf = gpd.read_file('01_유동인구데이터(SK텔레콤)/4개지역_행정동.DBF')
# prj = gpd.read_file('01_유동인구데이터(SK텔레콤)/4개지역_행정동.PRJ')
shp = gpd.read_file('01_유동인구데이터(SK텔레콤)/4개지역_행정동.SHP')
shx = gpd.read_file('01_유동인구데이터(SK텔레콤)/4개지역_행정동.SHX')
```

```python
seoul = dbf[dbf.SIDO_NM == "서울특별시"]
seoul
```

```python
# 동
dong_nowon = seoul[seoul.SGNG_NM == "노원구"].HDONG_NM.unique()
dong_junggu = seoul[seoul.SGNG_NM == "중구"].HDONG_NM.unique()

# 지도
geo_nowon = seoul[seoul.SGNG_NM == "노원구"][['HDONG_NM', 'geometry']]
geo_junggu = seoul[seoul.SGNG_NM == "중구"][['HDONG_NM', 'geometry']]
```

```python
time_202005 = pd.read_csv("C:/bg/bigcon/01_유동인구데이터(SK텔레콤)/4개지역_FLOW_TIME_202005.CSV", sep = "|")
time_202005
```

```python
time_202005_nowon = time_202005[time_202005.HDONG_NM.apply(lambda x: x in dong_nowon)]
time_202005_junggu = time_202005[time_202005.HDONG_NM.apply(lambda x: x in dong_junggu)]
```

```python
time_202005_nowon
```

```python
time_202005_nowon = pd.merge(gpd.GeoDataFrame(time_202005_nowon), geo_nowon, on = "HDONG_NM")
time_202005_nowon
```

```python
times = time_202005.columns[4:]
```

```python
ims = []
for time in times:
    ims.append([time_202005_nowon.plot(column = time, legend = "best").set_title(time)])
```
