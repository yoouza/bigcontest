### 유동인구 데이터 Visualizing

```
# 새로 추가한 geojson 데이터 (seoul-dong.geojson)
data_load = gpd.read_file('/content/drive/My Drive/project/Bigcon/2020빅콘테스트 문제데이터(혁신아이디어분야)/01_유동인구데이터(SK텔레콤)/seoul-dong.geojson')
data_seoul = data_load[data_load['시']=='서울특별시']
data_seoul

# 중구와 노원구만 우선 추출
data_no_joong = data_seoul[(data_seoul['구']=='중구') | (data_seoul['구']=='노원구')].reset_index(drop=True)
data_no_joong = data_no_joong[['행정동코드', '동','geometry']]
data_no_joong

data_no_joong.to_file("geo_data.geojson", driver="GeoJSON")
data_no_joong
```

```
# 빅콘 geo_data path
seoul_geo_path = 'geo_data.geojson'
seoul_geo = gpd.read_file(seoul_geo_path)
seoul_geo
```

```
### 분석 데이터 호출
# 4개지역_FLOW_AGE_201902.CSV 데이터 불러오기
float_pop = pd.read_csv('/content/drive/My Drive/project/Bigcon/2020빅콘테스트 문제데이터(혁신아이디어분야)/01_유동인구데이터(SK텔레콤)/4개지역_FLOW_AGE_201902.CSV', sep='|')
float_pop
```

```
# 20190201 & 1014 데이터로만 한정
float_pop_1 = float_pop[float_pop['STD_YMD']==20190201][['HDONG_CD','HDONG_NM', 'MAN_FLOW_POP_CNT_1014']].reset_index(drop=True)
float_pop_1.head(50)
```

```
# 서울시 중심부 좌표
center = [37.610, 126.985]
```

```
import folium
m = folium.Map(location=center, zoom_start=12)
```

```
folium.Choropleth(
    geo_data = seoul_geo_path,
    name='choropleth',
    data = float_pop_1, 
    columns=['HDONG_NM', 'MAN_FLOW_POP_CNT_1014'], 
    key_on='feature.properties.동',

    # 색 옵션 
    fill_color='Reds', 
    legend_name='1014 인구유동', 
).add_to(m)
folium.LayerControl().add_to(m)
```

```
folium.LayerControl().add_to(m)
```

```
# 지도 출력
m.save('map.html')
m
```
