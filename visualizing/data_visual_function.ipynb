## 특정날짜 시간별 인구 folium 그린 후 저장하는 함수

```python
def draw_folium(yy_mm, dd, h, path):
    
    df = globals()['time_{}'.format(yy_mm)]
    yy_mm_dd = yy_mm*100 + dd
    
    center = [37.610, 126.985]
    m = folium.Map(location=center, zoom_start=12)
    
    folium.Choropleth(
        geo_data = seoul_geo_path,
        name='choropleth',
        data = df[['HDONG_CD', 'HDONG_NM', times[h]]], 
        columns=['HDONG_NM', times[h]], 
        key_on='feature.properties.동',
        fill_color='Reds',
        legend_name='{0} {1}시 인구유동'.format(yy_mm_dd, h),
    ).add_to(m)
    folium.LayerControl().add_to(m)
    
    m.save(path + '{0}_{1}.html'.format(yy_mm_dd, h))
```
